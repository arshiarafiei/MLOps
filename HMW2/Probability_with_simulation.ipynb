```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mu = 5
x = range(0,15)
y = stats.poisson.pmf(x, mu)
```

```python
# probability that Poisson distribution outcome, mean 5, is equal to 4
print(stats.poisson.pmf(4,5));

# probability that Poisson distribution outcome, mean 5, is at least 8
print(1 - stats.poisson.cdf(7, 5))
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
sim = stats.poisson.rvs(5, size = 10000000)
```

```python
plt.hist(sim, bins = 22, facecolor='blue')
plt.show()
```

```python
# normal distribution example
# mean 3, standard deviation 1
# probability that outcome is greater than 4
print(1 - stats.norm.cdf(4, 3, 1))
```

```python
# uniform distribution
# bounded by 1 and 5
# conditional probability that outcome is less than 2, given that it's less than 3
# P(A|B) = P(A and B) / P(B)
print(stats.uniform.cdf(2,1,5)/stats.uniform.cdf(3,1,5))
```

```python
# quantile function
# mean 3, standard deviation 1
# 0.3 quantile
print(stats.norm.ppf(0.3, 3, 1))
```

```python
# Simulation example
# 2 unlinked events sequentially
sim1, sim2 = stats.norm.rvs(100, 20, 100000), stats.norm.rvs(40, 20, 100000)
sim = sim1 + sim2
plt.hist(sim, bins = 22, facecolor='blue')
plt.show()
```

```python
print(np.mean(sim), np.var(sim), np.std(sim))
```

```python
# 2 distinct events happening in parallel
sim1 = stats.norm.rvs(50, 20, 100000)
sim2 = stats.norm.rvs(40, 20, 100000)
sim = np.amax(np.column_stack((sim1,sim2)),1)
plt.hist(sim, bins = 22, facecolor='blue')
plt.show()
```

```python
print(np.mean(sim), np.var(sim), np.std(sim))
```

```python
# 95% 2-sided Confidence Interval
np.quantile(sim,(0.025,0.975))
```

```python
# 95% 1-sided Confidence Interval (only care about delays)
np.quantile(sim,0.95)
```
